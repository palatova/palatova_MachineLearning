```python
#пункты 3-5
import pandas
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

data_for_heroes = pandas.read_csv('./features.csv', index_col='match_id')
target=data_for_heroes['radiant_win']

del data_for_heroes['duration']
del data_for_heroes['radiant_win']
del data_for_heroes['tower_status_radiant']
del data_for_heroes['tower_status_dire']
del data_for_heroes['barracks_status_radiant']
del data_for_heroes['barracks_status_dire']

#количество идентификаторов героев
count_heroes=data_for_heroes['d3_hero'].unique()
print(len(count_heroes))

#мешок слов
N=len(count_heroes)
print(len(data_for_heroes['r1_hero'].unique()))
X_pick = np.zeros((data_for_heroes.shape[0], N+4))

for i, match_id in enumerate(data_for_heroes.index):
    for p in range(5):
        X_pick[i, data_for_heroes.ix[match_id, 'r%d_hero' % (p+1)]-1] = 1
        X_pick[i, data_for_heroes.ix[match_id, 'd%d_hero' % (p+1)]-1] = -1
        
X_pick = pandas.DataFrame(X_pick, index=data_for_heroes.index)
        
data_for_heroes.drop(["lobby_type","r1_hero","r2_hero","r3_hero","r4_hero","r5_hero","d1_hero","d2_hero","d3_hero","d4_hero","d5_hero"], axis=1, inplace=True)
data_train = pandas.concat([data_for_heroes, X_pick], axis=1)
data_train=data_train.fillna(data_train.mean())

scaler = StandardScaler()
priznaki_train_scaled = scaler.fit_transform(data_train)


clf = LogisticRegression(random_state=241, penalty='l2').fit(priznaki_train_scaled, target)
pred = clf.predict_proba(priznaki_train_scaled)[:, 1]

grid = {'C': np.power(10.0, np.arange(-5, 5))}
cv = KFold(n_splits=5, shuffle=True, random_state=241)
gs = GridSearchCV(clf, grid, scoring='roc_auc', cv=cv)
gs.fit(priznaki_train_scaled, target)
C_best=gs.best_estimator_.C
print("Здесь лучший параметр C №3")
print(C_best)

clf_better=LogisticRegression(random_state=241, penalty='l2',C=C_best).fit(priznaki_train_scaled, target)
pred = clf_better.predict_proba(priznaki_train_scaled)[:, 1]

start_time = datetime.datetime.now()
cross_val = cross_val_score(clf_better,priznaki_train_scaled, target, cv=cv,scoring='roc_auc')
print ('Time elapsed:', datetime.datetime.now() - start_time)

print("А здесь лучшее значение  кросс валидации при лучшем C №3")
print(cross_val.mean())
```
